```python
init = 0
limit = 778
#limit = 8803
testes = ['5', '10', '15', '20', '25']
path = "/home/anderson/trabalhos/temp/cap52/early_warning.csv"
early_warning = pd.read_csv(path, sep=";")
y_origem = early_warning["has_bot"][init:limit]
y_origem = y_origem.fillna(0)
for i in testes:
    cols= ["total_pacotes","total_pacotes_icmp","total_pacotes_udp","total_pacotes_tcp","maior_pacote","menor_pacote","soma_pacotes","total_ips_origem","total_ips_destino"]
    colunas = [i+'_entropia_'+j for j in cols]
    colunas.extend([i+'_kurt_'+j for j in cols])
    colunas.extend([i+'_skw_'+j for j in cols])

    X = early_warning[colunas].copy()[init:limit]
    X = X.fillna(0)
    #y = early_warning["has_bot_origem_destino"][init:limit]
    #y_origem = early_warning["has_bot"][init:limit]

    num_features = len(X.columns)

    bestfeatures = SelectKBest(k=num_features)

    fit = bestfeatures.fit(X, y_origem)
    dfscores = pd.DataFrame(fit.scores_)
    dfcolumns = pd.DataFrame(X.columns)

    featureScores = pd.concat([dfcolumns,dfscores],axis=1)
    featureScores.columns = ['Specs','Score']  
    scores = featureScores.nlargest(11, 'Score' )

    plt.title(f"Melhores features (Univariate Selection)")
    plt.xlabel("Score")
    plt.barh(scores["Specs"], scores["Score"])
    plt.show()
    
    
```
