# Transformar CSV em PCAP
# Esse código lê todos os pcaps de uma pasta e extrai alguns dados para criar um CSV com todos para cada PCAP

```python
#configurações do script
origem_pacotes = "/home/anderson/trabalhos/temp/cap52/ieee-acess/quebrado"
destino_csvs = "/home/anderson/trabalhos/temp/cap52/ieee-acess/csvs/"
pcap_key = "###PCAP###"
csv_out_key = "###OUT###"
num_processos = 3
lista_comandos = []
comando = "tshark -o gui.column.format:\"SP,%uS,DP,%uD,SA,%s,DA,%d,PR,%p\" -r "+pcap_key+" -T fields -e frame.time -e frame.time_epoch -e  frame.time_delta -e frame.len -e frame.cap_len  -e eth.type -e eth.src -e eth.dst 	-e ip.version -e _ws.col.SA -e _ws.col.DA -e _ws.col.PR -e ip.ttl -e ip.checksum -e _ws.col.SP -e _ws.col.DP  -e tcp.flags -e tcp.flags.fin -e tcp.flags.syn -e tcp.flags.reset -e tcp.flags.push -e tcp.flags.ack -e tcp.flags.urg   -e tcp.window_size_value  -e tcp.seq -e tcp.ack_raw -E header=y -E separator=\";\" >"+csv_out_key+".csv"
```

```python
#DEFINIR IMPORTS
import os
import pandas as pd
import multiprocessing
```

```python
#MÉTODO PARA LER OS PACOTES DE UMA PASTA
def ler_pacotes_pasta():
    return sorted(os.listdir(origem_pacotes))
```

```python
def executar_comando(executar):
    stream = os.popen(executar)
    #return int(str(stream.read()).replace("\n", ""))
    return str(stream.read()).replace("\n", "")
```

```python
pacotes = ler_pacotes_pasta()

for pacote in pacotes:
    lista_comandos.append(comando.replace(pcap_key,origem_pacotes+"/"+pacote).replace(csv_out_key,destino_csvs+pacote))

print(lista_comandos[0])
    
```

```python
pool = multiprocessing.Pool(processes=num_processos)
pool.map(executar_comando,lista_comandos)
```
